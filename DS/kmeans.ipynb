# K Means Clustering

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
data, _ = make_blobs(n_samples=300)
```

```python
model = KMeans(random_state=42, n_clusters=5, n_init='auto')
model.fit(data)
```

```python
labels = model.labels_

# Stores the coordinates of the centroids for these 5 clusters
centroids = model.cluster_centers_
```

```python
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', marker='o', alpha=0.6)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('K-Means')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend(list(labels))
plt.show()
```

```python
model.predict([[2, -4], [6, 6], [6, -2]])
```
