# Decision Tree

A Decision Tree is a supervised machine learning algorithm used for both classification and regression tasks. It models decisions based on a tree-like structure where **each internal node represents a feature** (attribute), **each branch represents a decision rule**, and **each leaf node represents an outcome** (class label or regression value).

### Structure of a Decision Tree
- **Root Node**: The top node that represents the entire dataset, which is split into two or more homogeneous sets.
- **Internal Nodes**: Nodes that represent tests on features and lead to further splits.
- **Leaf Nodes**: Terminal nodes that represent the final output (class label in classification or a continuous value in regression).

## Algorithm
1. **Splitting**: The process of dividing a node into two or more sub-nodes. The best split is determined using metrics like **Gini impurity**, entropy (information gain), or mean squared error (for regression).
  
2. **Decision Rules**: Each internal node corresponds to a feature and a threshold value that decides how to split the data. For example, if a feature is "Age" and the threshold is 30, the split will separate instances where "Age <= 30" from those where "Age > 30".

3. **Stopping Criteria**: The algorithm stops splitting when:
   - A maximum depth is reached.
   - A minimum number of samples in a node is reached.
   - No further information gain can be achieved.

4. **Pruning**: After building the tree, pruning may be performed to reduce overfitting. This involves removing nodes that have little importance, which simplifies the model and enhances its generalization to unseen data.

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
dataset = load_iris()

X, y = dataset.data, dataset.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
classifier = DecisionTreeClassifier()
```

```python
classifier.fit(X_train, y_train)
```

```python
y_predict = classifier.predict(X_test)

accuracy_score(y_test, y_predict)
```

```python
result = classifier.predict([[4, 3, 1, 0]])
print(dataset.target_names[result])
```

```python
tree.plot_tree(classifier)
```

```python
import matplotlib.pyplot as plt

print(y_test, y_predict)
plt.scatter(y_test, y_predict, alpha=0.7, color='blue')
plt.title('Actual vs Predicted Values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.grid(True)
plt.axis('equal')
plt.show()
```
