# Matplotlib

Matplotlib is a low level graph plotting library in python that serves as a visualization utility.
Its numerical mathematics extension NumPy.
Most of the Matplotlib utilities lies under the `pyplot` submodule, and are usually imported under the `plt` alias.

```python
import matplotlib.pyplot as plt
```

#### Plotting Lines

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Plotting the lines connecting the points (xi, yi)
plt.plot(x, y, color='g', linestyle='--', linewidth=3, marker='x', markersize=15, label='My Line')

# Title and Labels
plt.title("Plotting Lines")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Legend is used to identify the elements in the plot
plt.legend()

plt.show()
```

```python
import numpy as np

x = np.arange(0, 10, 1)
y = 3 * x + 5

plt.plot(x, y, color='red', marker='o', markersize='5', label='graph')
plt.legend()
plt.show()
```

#### Sine and Cosine

```python
x = np.linspace(-np.pi, np.pi, 100) # 100 points between -pi and pi
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='Sine', color='blue')
plt.plot(x, y2, label='Cosine', color='green')
plt.legend()
plt.show()
```

#### Subplots

Subplots allow us to create multiple plots in a single figure.

```python
# Create a figure
plt.figure()

plt.subplot(2, 1, 1) # subplot(rows, cols, current_plot)

plt.plot(x, y1)
plt.title("Sine Wave")

plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title("Cosine Wave")

plt.tight_layout() # Adjust spacing to prevent overlap
plt.show()
```

#### Bar Chart

```python
x = np.arange(1, 6)
y = np.random.randint(1, 50, 5)

plt.bar(x, y, align='center')
plt.show()
```

```python
plt.title("Red bars")
plt.bar(np.arange(1, 6), np.random.randint(1, 10, 5), color='r')
plt.show()
```

#### Pie Chart

```python
x = np.array([4, 4, 1])

plt.pie(x, colors=['green', 'blue', 'orange'], labels=['A', 'B', 'C'], autopct="%1.1f%%", explode=(0, 0, 0.1), startangle=90)
plt.show()
```
